```python
import pandas as pd
```

```python
web_text = pd.read_csv("web-pages.csv")
```

```python
web_list = web_text['domain'].unique()
web_list
```

```python
domain_list = pd.DataFrame(columns={'domain', 'length'})

for i in range(0, len(web_list)):
    new_df = web_text[web_text['domain'] == web_list[i]]
    domain_list.loc[i, 'domain'] = web_list[i]
    domain_list.loc[i, 'length'] = len(new_df)
    
    #domain = web_list[i].replace('.', '_')
    #new_df.to_csv('Desktop/RESEARCH/' +domain+ '.csv')
    #print(i, '\t', web_list[i], '\t', len(new_df))
```

```python
domain_list = domain_list.sort_values(by=['length'], ascending=False)
domain_list = domain_list[domain_list['length'] > 10].reset_index()
domain_list
```

```python
for i in range(0, len(domain_list)):
    domain = domain_list.loc[i, 'domain']
    file_name = domain.replace('.', '_')
    new_df = web_text[web_text['domain'] == domain]
    new_df.to_csv('OneDrive - Brock University/SCHOOL/YEAR 3/RESEARCH_2/' +file_name+ '.csv')
    print(i, '\t', domain, '\t', len(new_df))
```
